```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def weighted_average(series,beta):
    v = [0]
    n = series.shape[0]
    for i in range(n):
        v.append(beta*v[-1] + (1-beta)*series.iloc[i])

    return v[1:]
```

```python

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.05)
f = np.exp(-np.square((x)-mean)/2*variance)/(np.sqrt(2*np.pi*variance))+np.random.uniform(-0.3,0.2,(len(x)))
plt.scatter(x,f,s=5)
```

```python
df = pd.DataFrame(f)
```

```python
plt.plot(x, weighted_average(df,0.2),linewidth=2,alpha=0.4,label='beta=0.2')
plt.plot(x, weighted_average(df,0.9),linewidth=2,label='beta=0.9')
plt.plot(x, weighted_average(df,0.98),linewidth=2,label='beta=0.98')
plt.scatter(x,f,s=19,c='k',alpha=0.6)
plt.xlabel('x',fontsize=20)
plt.ylabel('F',fontsize=20)
plt.legend()
```
